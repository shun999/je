```python
# df["Time_test"] = pd.to_timedelta(df["Time"])

df['Time_test'] = pd.to_timedelta(df['Time'].str.replace(r'\.(\d+)', r':\1', regex=True))
df['Time_test'] = df['Time_test'] % pd.Timedelta(days=1)
df['Time_test'] = df['Time_test'] + pd.to_timedelta((df.groupby('Time').cumcount() * 0.2), unit='s')
```

```python
# df["Time_test"] = df["Time_test"].dt.total_seconds()
```

```python
df["Time_test"] = df["Time_test"] - df["Time_test"].iloc[0]
```

```python
df["Time_test"] = pd.to_timedelta(df["Time_test"], unit='s')
```

```python
df["Time_test"][:10]
```

```python
df = df.set_index("Time_test")
df = df.drop(columns=["Time"])
```
