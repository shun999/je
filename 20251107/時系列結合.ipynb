# 被験者ごと分割されている時系列データを結合

```python
import pandas as pd
from pathlib import Path
```

```python
base_dir = Path("E:/Je respire/解析/深層学習/2. タイムスタンプ作成")
```

```python
files = list(base_dir.glob("*bvpsc.csv"))
```

```python
files
```

```python
df = [pd.read_csv(file) for file in files]
```

```python
df[2]
```

```python
df_all = pd.concat([df[i] for i in range(len(df)) if i not in (0, 1) ], ignore_index=True)
df_all = pd.concat([df_all, df[0]], ignore_index=True)
df_all = pd.concat([df_all, df[1]], ignore_index=True)
```

```python
df_all
```

```python
df_all.to_csv("E:/Je respire/解析/深層学習/3. 時系列結合/bvpsc.csv", index=False)
```
