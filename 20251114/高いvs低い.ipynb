```python
import pandas as pd
```

```python
df = pd.read_excel(r"E:\Je respire\解析\1.3 脳血流と問題カテゴリの関連 - 背外側追加 - メンタル指標追加.xlsx", sheet_name="ディアビアイ")
```

```python
df_high_all = df[df["subject"].isin([1, 2, 5, 6, 9])]
df_low_all = df[df["subject"].isin([3, 4, 7, 10, 11])]
```

```python
df_high_all = df_high_all[df_high_all["q_category"]==3].reset_index(drop=True)
df_low_all = df_low_all[df_low_all["q_category"]==3].reset_index(drop=True)
```

```python
df_high_all.columns
```

```python
for i, column in enumerate(df_high_all.columns):
    target = column
    df_compare = pd.DataFrame({"High": df_high_all[target],
                            "Low": df_low_all[target]})

    import matplotlib.pyplot as plt
    import seaborn as sns

    plt.rcParams['font.family'] = 'MS Gothic'  # または 'Meiryo'

    sns.boxplot(data=df_compare, palette="Set3")
    plt.title(f'High vs Low ')
    plt.ylabel("score")
    plt.xlabel(target)
    plt.show()

    df_compare=df_compare.dropna()

    from scipy.stats import brunnermunzel

    t_stat, p_value = brunnermunzel(df_compare["High"], df_compare["Low"])

    print("t値 :", t_stat)
    print("p値 :", p_value)
```

```python
from scipy.stats import permutation_test
import numpy as np

for i, column in enumerate(df_high_all.columns):
    target = column
    df_compare = pd.DataFrame({"High": df_high_all[target],
                            "Low": df_low_all[target]})

    import matplotlib.pyplot as plt
    import seaborn as sns

    plt.rcParams['font.family'] = 'MS Gothic'  # または 'Meiryo'

    sns.boxplot(data=df_compare, palette="Set3")
    plt.title(f'High vs Low ')
    plt.ylabel("score")
    plt.xlabel(target)
    plt.show()

    df_compare=df_compare.dropna()

    def statistic(a, b):
        return a.mean() - b.mean()

    result = permutation_test(
        (df_compare["High"], df_compare["Low"]),
        statistic,
        n_resamples=5000,
        alternative='two-sided'
    )

    print(result.statistic)
    print(result.pvalue)
```
