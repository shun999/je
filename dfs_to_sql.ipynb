```python
import pandas as pd
import sqlalchemy as sa
```

```python
#公開ユーザー名はopen
import getpass
host="192.168.100.68" #サーバーのIPアドレス
user="open"
password=getpass.getpass("Enter your password:")
database = "jerespire" #データベース名
port = "3306"
```

```python
url = f'mysql+pymysql://{user}:{password}@{host}:{port}/{database}?charset=utf8'
engine = sa.create_engine(url, echo=False)
```

```python
from pathlib import Path

folder = Path(r"E:\Je respire\生体情報\4. 標準化")

file_list = [str(path) for path in folder.rglob("*.*.xlsx")]  # すべてのファイル
print(file_list)
```

```python
import os

names = [os.path.splitext(os.path.basename(p))[0] for p in file_list]
print(names)
```

```python
from tqdm import tqdm
import time

for k, path in enumerate(file_list):
    for i in tqdm(range(33)):
        df = pd.read_excel(path, sheet_name=f"Q{i+1}")
        df.to_sql(name=f"raw_std_{names[k]}_Q{i+1}", con=engine, index=False)
        time.sleep(0.1)
```
