```python
from pathlib import Path

folder = Path(r"E:\Je respire\生体情報\4. 標準化")

file_list = [str(path) for path in folder.rglob("*.*.xlsx")]  # すべてのファイル
print(file_list)
```

```python
import os

names = [os.path.splitext(os.path.basename(p))[0] for p in file_list]
print(names)
```

```python
from tqdm import tqdm
import time

for k, path in enumerate(file_list):
    for i in tqdm(range(33)):
        df = pd.read_excel(path, sheet_name=f"Q{i+1}")
        df.to_sql(name=f"raw_std_{names[k]}_Q{i+1}", con=engine, index=False)
        time.sleep(0.1)
```

```python
from tqdm import tqdm
import pandas as pd
import time
for k, path in enumerate(file_list):

    for i in tqdm(range(33)):
        df = pd.read_excel(path, sheet_name=f"Q{i+1}")

        # 1️⃣ カラム名を統一（小文字・空白除去・ピリオド置換）
        df.columns = (
            df.columns
            .str.replace(r'\.', '_', regex=True)
            .str.replace(r'\s+', '_', regex=True)
            .str.strip()
            .str.lower()
        )

        # 2️⃣ 大文字小文字無視での重複除去
        new_cols = []
        seen_lower = {}
        for col in df.columns:
            col_lower = col.lower()
            if col_lower in seen_lower:
                seen_lower[col_lower] += 1
                new_cols.append(f"{col}_{seen_lower[col_lower]}dup")
            else:
                seen_lower[col_lower] = 0
                new_cols.append(col)
        df.columns = new_cols

        # 3️⃣ 重複チェック（デバッグログ）
        dups = df.columns[df.columns.duplicated()]
        if len(dups) > 0:
            print(f"⚠️ 重複検出: {dups.tolist()}")

        # 4️⃣ SQLに保存（既存テーブルは置換）
        df.to_sql(
            name=f"raw_std_{names[k]}_q{i+1}",
            con=engine,
            index=False,
            if_exists='replace'
        )

        time.sleep(0.1)

```
