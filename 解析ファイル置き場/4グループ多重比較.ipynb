```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_excel("E:/Je respire/解析/4グループ多重比較.xlsx", sheet_name="認知")
df2 = pd.read_excel("E:/Je respire/解析/4グループ多重比較.xlsx", sheet_name="読解")
df3 = pd.read_excel("E:/Je respire/解析/4グループ多重比較.xlsx", sheet_name="記憶")
df4 = pd.read_excel("E:/Je respire/解析/4グループ多重比較.xlsx", sheet_name="集中注意")
```

```python
df1 = df1.iloc[:, np.r_[0, 6:28]]
df2 = df2.iloc[:, np.r_[0, 6:28]]
df3 = df3.iloc[:, np.r_[0, 6:28]]
df4 = df4.iloc[:, np.r_[0, 6:28]]
```

```python
df1_mean = df1.groupby("subject").mean()
df2_mean = df2.groupby("subject").mean()
df3_mean = df3.groupby("subject").mean()
df4_mean = df4.groupby("subject").mean()
```

```python
df1_mean.columns = [f"{col}_recognition" for col in df1_mean.columns]
df2_mean.columns = [f"{col}_reading" for col in df2_mean.columns]
df3_mean.columns = [f"{col}_memory" for col in df3_mean.columns]
df4_mean.columns = [f"{col}_attention" for col in df4_mean.columns]
```

```python
df_mean = pd.concat([df1_mean, df2_mean, df3_mean, df4_mean], axis=1)
```

```python
df_chs = [
    df_ch1, df_ch2, df_ch3, df_ch4, df_ch5, df_ch6,
    df_ch7, df_ch8, df_ch9, df_ch10, df_ch11, df_ch12,
    df_ch13, df_ch14, df_ch15, df_ch16, df_ch17, df_ch18,
    df_ch19, df_ch20, df_ch21, df_ch22
]

import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'MS Gothic'  # または 'Meiryo'

for i, df_ch in enumerate(df_chs, start=1):
    sns.boxplot(data=df_ch)
    plt.title(f'CH{i} 箱ひげ図')
    plt.ylabel('脳血流賦活量')
    plt.xlabel('問題の種類')
    plt.show()
```

```python
# 'df' はあなたの元のデータフレームとします
# 列名から不要な部分を削除して 'task' 名だけにします
df_ch1.columns = ['recognition', 'reading', 'memory', 'attention']

# 被験者IDを追加
df_ch1['subject'] = range(len(df_ch1))

# pd.melt()でロングフォーマットへ変換
df_ch1_long = pd.melt(df_ch1, id_vars='subject', var_name='task', value_name='value')

# 結果の確認
print(df_ch1_long.head())
```

```python
import pingouin as pg

# 一元配置反復測定分散分析を実行
# withinに指定する要因が'task'のみになる
aov = pg.rm_anova(dv='value', 
                  within='task', 
                  subject='subject', 
                  data=df_ch1_long)

# 結果を表示
print(aov)
```

```python
dfs = [df_ch1, df_ch2, df_ch3, df_ch4, df_ch5, df_ch6, df_ch7, df_ch8, df_ch9, df_ch10,
        df_ch11, df_ch12, df_ch13, df_ch14, df_ch15, df_ch16, df_ch17, df_ch18, df_ch19, df_ch20,
        df_ch21, df_ch22]
for i, df in enumerate(dfs, start=1):
    # 'df' はあなたの元のデータフレームとします
    # 列名から不要な部分を削除して 'task' 名だけにします
    df.columns = ['recognition', 'reading', 'memory', 'attention']

    # 被験者IDを追加
    df['subject'] = range(len(df))

    # pd.melt()でロングフォーマットへ変換
    df_long = pd.melt(df, id_vars='subject', var_name='task', value_name='value')

    # 結果の確認
    print(df_long.head())

    import pingouin as pg

    # 一元配置反復測定分散分析を実行
    # withinに指定する要因が'task'のみになる
    aov = pg.rm_anova(dv='value', 
                    within='task', 
                    subject='subject', 
                    data=df_long)

    # 結果を表示
    print(f"{i}チャンネルの結果：{aov}")
```
