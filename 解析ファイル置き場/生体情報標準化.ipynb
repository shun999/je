```python
dfs = [df2, df3, df4, df5, df6, df7, df8, df9]
```

```python
for i, dataset in enumerate(dfs):
    outputname = f"standardized_output{i+2}.xlsx"

    merged_df = pd.concat(
        [df.assign(Quarter=sheet_name) for sheet_name, df in dataset.items() if sheet_name.startswith("Q")],
        ignore_index=True
    )

    columns = ["left_pupil", "right_pupil", 
            "CH1", "CH2", "CH3", "CH4", "CH5", "CH6", "CH7", "CH8", "CH9", "CH10", "CH11", "CH12", "CH13", "CH14", "CH15", "CH16", "CH17", "CH18", "CH19", "CH20", "CH21", "CH22",
            "CH1.1", "CH2.1", "CH3.1", "CH4.1", "CH5.1", "CH6.1", "CH7.1", "CH8.1", "CH9.1", "CH10.1", "CH11.1", "CH12.1", "CH13.1", "CH14.1", "CH15.1", "CH16.1", "CH17.1", "CH18.1", "CH19.1", "CH20.1", "CH21.1", "CH22.1",
            "CH1.2", "CH2.2", "CH3.2", "CH4.2", "CH5.2", "CH6.2", "CH7.2", "CH8.2", "CH9.2", "CH10.2", "CH11.2", "CH12.2", "CH13.2", "CH14.2", "CH15.2", "CH16.2", "CH17.2", "CH18.2", "CH19.2", "CH20.2", "CH21.2", "CH22.2",
            "Sensor-E:SC/GSR", "Sensor-G:BVP", "[G] BVP Amp.", "[G] Heart Rate", "[G] HRV Amp.", "[G] HRV-LF Power (0,04-0,16 Hz)", "[G] HRV-HF Power (0,16-0,4 Hz)", "[G] HRV-LF / HRV-HF "
            ]
    means = {}
    stds = {}

    for  col in (columns):
        means[col] = merged_df[col].mean()
        stds[col] = merged_df[col].std()

    print(means)
    print(stds)

    standardized_dfs = {}

    for sheet_name, df in dataset.items():

            for col in columns:
                if col not in df.columns:
                    continue
                
                # 標準化（Zスコア）
                mean = means[col]
                std = stds[col]
                df[f"{col}_std"] = (df[col] - mean) / std

                # 加工済みの DataFrame を辞書に格納
                standardized_dfs[sheet_name] = df

    with pd.ExcelWriter(outputname) as writer:
        for sheet_name, df in standardized_dfs.items():
            df.to_excel(writer, sheet_name=sheet_name, index=False)

    print(f"{i+2}番目のデータセットの標準化が完了しました。出力ファイル名: {outputname}")
```
